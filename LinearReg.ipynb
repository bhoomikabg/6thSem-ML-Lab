```python
import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split

# 1️⃣ Load dataset
data = fetch_california_housing()
X = data.data
y = data.target
feature_names = data.feature_names

# 2️⃣ Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# 3️⃣ Add bias column
ones = np.ones((X_train.shape[0], 1))
X_train_b = np.hstack((ones, X_train))

# 4️⃣ Compute coefficients using PSEUDO-INVERSE (more stable)
beta = np.linalg.pinv(X_train_b).dot(y_train)

print("Model trained successfully!\n")

# 5️⃣ Show valid feature ranges
print("Enter values within these ranges:\n")
for i, feature in enumerate(feature_names):
    print(f"{feature}: Min = {X[:, i].min():.2f}, Max = {X[:, i].max():.2f}")

print("\nEnter the following values:")

# 6️⃣ Take user input
user_input = []
for feature in feature_names:
    value = float(input(f"{feature}: "))
    user_input.append(value)

user_input = np.array(user_input)

# 7️⃣ Add bias term
user_input_b = np.hstack(([1], user_input))

# 8️⃣ Predict
predicted_value = user_input_b.dot(beta)

# Prevent negative prediction
predicted_value = max(0, predicted_value)

print("\nPredicted House Value (in $100,000s):", predicted_value)
print("Predicted House Value in dollars: $", predicted_value * 100000)
```
